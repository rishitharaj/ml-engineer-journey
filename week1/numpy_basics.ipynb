```python
# Import & Setup

import numpy as np
np.random.seed(42)
```

```python
##Arrays

# 1D Array
a = np.array([1, 2, 3, 4])
print("1D Array:", a)

# 2D Array
b = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", b)

# Zeros and Ones
zeros = np.zeros((2, 3))
ones = np.ones((3, 3))

print("Zeros:\n", zeros)
print("Ones:\n", ones)
```

```python
# Array Operations

x = np.array([1, 2, 3])
y = np.array([4, 5, 6])


print("Addition:", x + y)
print("Multiplication:", x * y)
print("Mean:", np.mean(x))
print("Standard Deviation:", np.std(x))
```

```python
# Matrix Multiplication (how neural networks compute forward pass.)

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Method 1
result1 = np.dot(A, B)

# Method 2
result2 = A @ B

print("Shape of A:", A.shape)
print("Shape of B:", B.shape)
print("Matrix Multiplication (dot):\n", result1)
print("Matrix Multiplication (@):\n", result2)
```

```python
# Matrix multiplication is used in neural networks to compute the weighted sum of inputs.
# Every forward pass in a neural network is essentially a series of matrix multiplications.
```

```python
# Broadcasting

a = np.array([1, 2, 3])
b = 2

print("Broadcasting result:", a * b)
```

```python
# Vectorization

arr = np.array([1, 2, 3, 4])

result = []
for i in arr:
    result.append(i * 2)

print("Loop result:", result)
```

```python
# Vectorized, fast

result_vec = arr * 2
print("Vectorized result:", result_vec)
```

```python
# Performance Comparison

import time

large_array = np.random.rand(1000000)

# Loop
start = time.time()
res_loop = [i * 2 for i in large_array]
print("Loop time:", time.time() - start)

# Vectorized
start = time.time()
res_vec = large_array * 2
print("Vectorized time:", time.time() - start)
```

```python
# What is Vectorization? 
# Vectorization allows operations on entire arrays without explicit loops, making computations faster and efficient.
```

```python
### Why Vectorization is Important

# Vectorization eliminates Python loops and uses optimized low-level implementations (C/C++).
# This makes computations significantly faster and scalable for large datasets.

# In machine learning, vectorization is critical because models operate on large datasets,
# and inefficient loops can slow down training drastically.
```

```python
# Mini ML Concept 

# Simulating a simple linear model: y = Wx + b

X = np.array([[1, 2], [3, 4]]) # features
W = np.array([[0.5], [1.0]]) # weight
b = 0.1 # bias

y = X @ W + b
print("Output:\n", y)
```
